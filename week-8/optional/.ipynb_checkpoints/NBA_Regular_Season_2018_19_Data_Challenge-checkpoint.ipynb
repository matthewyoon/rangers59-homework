# What was the average age of player in the league?

```python
avg_age = df['Age']
print(round(avg_age.mean(),2))
```

# What player scored the most points?

```python
most_points = df.groupby('Name')[['Points']].sum().sort_values('Points',ascending=False)
most_points.head(3)
```

# What player had the most blocks during the season? Was it a post player (F/C)?

```python
most_blocks = df.groupby(['Name'])[['Blocks']].sum().sort_values('Blocks',ascending=False)
most_blocks
# Myles Turner and yes he is a Center
```

# What player had the best 3-pt percentage? 

```python
threes = df[['Name','FG3%']]
threes = threes.dropna(how='any')
threes = threes.sort_values('FG3%',ascending=False)
threes.head(10)
```

```python
# Would want players that have attempted at least a certain amount of 3's per game instead of by pure percentage
actual_threes = df[['Name','FG3%','FG3A']]
actual_threes = actual_threes.dropna(how='any')
actual_threes = actual_threes[actual_threes['FG3A']>2]
actual_threes = actual_threes.sort_values('FG3%',ascending=False)
actual_threes.head()
```

# Who played the most minutes during the season

```python
most_minutes = df.groupby('Name')[['MP']].sum().sort_values('MP',ascending=False)
most_minutes.head()
```

# What player given their player effiecency rating was the clutchest during the season?

```python
clutch = df.groupby('Name')[['PER']].max().sort_values('PER',ascending=False)
clutch.head()
```

# What team had the youngest roster?

```python
youngest = df.groupby('Team')[['Age']].mean().sort_values('Age')
youngest.head()
```

# Who is the highest paid player during the seasion?

```python
cash = df[['Name','Salary']]
# Get names of indexes for which column Age has value 30
cash_index = df[df['Salary'] == '-' ].index
# Delete these row indexes from dataFrame
cash.drop(cash_index , inplace=True)
cash['Salary'] = cash['Salary'].astype(int)
cash = cash.sort_values('Salary',ascending=False)
cash.head(5)
```
